```python
import socket
import pickle
import pandas as pd
import time
arr = []
def read_data_from_csv(csv_file):
    # Read data from the CSV file
    df = pd.read_csv(csv_file)

    # Convert the data to a list of dictionaries (each dictionary represents a row)
    data_list = df.to_dict('records')

    return data_list

def send_data_to_fog(data_list):
    HOST = '172.20.10.6'  # Fog node IP address or '0.0.0.0' to accept connections from any IP
    PORT = 8888        # Fog node port number

    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((HOST, PORT))

        for data in data_list:
            data_bytes = pickle.dumps(data)
            t1 = time.time()
            s.sendall(data_bytes)
            # Receive the result from the fog node
            result_bytes = s.recv(4096)
            t2 = time.time()
            arr.append(t2-t1)
            result = pickle.loads(result_bytes)
            # Process the result (e.g., print it)
            print(data)
            print("Result:", result)

if __name__ == "__main__":
    # Replace 'your_dataset_file.csv' with the actual CSV file path
    csv_file = "edge data.csv"

    # Read data from the CSV file
    data_list = read_data_from_csv(csv_file)

    # Send the data to the fog node for processing
    send_data_to_fog(data_list)
    print("\n\n\n")
    print("The Average Latency is:", end=" ")
    print((sum(arr)/len(arr)))
    arr.sort()
    print("the top 5 lowest latecies are: ", end=" ")
    print(arr[:5])
```
